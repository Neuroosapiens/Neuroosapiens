#  IMPORTING MATLAB DATASET

```python
#Importing the cvs file using Pandas
import pandas as pd
#Importing Matplotlib to generate graphs
import matplotlib.pyplot as plt
# Setting up Data Frame from the cvs file
df = pd.read_csv(r'Lab0_data.csv')
```

# GRAPHS OF DATA

## Time Vs Vehicle Speed

```python
# Setting up size of graph
plt.figure(figsize=(20, 11))
# Setting up the range and divisions in graph
plt.xticks(range(0, 850, 30))
# Displalying a grid in the background
plt.grid()
# Setting up the Width of the Data Line
plt.rcParams['lines.linewidth'] = 5
# Title of the Graph and Size of Text
plt.title('Time Vs Vehicle Speed', size=(50))
# Labeling and Size text Size of X axis
plt.xlabel('Time [s]', size=(25))
# Labeling and text size of the Y axis
plt.ylabel('Speed [mph]', size=(25))
# Plotting the Graph
plt.plot('time_sec', 'VehicleSpeed_mph', data=df)
```

# TIME Vs SPaT SIGNAL

```python
df_red = df[df.SPaT_signal == 1]
```

```python
df_yellow = df[df.SPaT_signal == 2]
```

```python
df_green = df[df.SPaT_signal == 3]
```

```python
plt.figure(figsize=(20, 11))
plt.xticks(range(0, 850, 50))
plt.grid()
plt.rcParams['lines.linewidth'] = 5
plt.title('Time Vs SPaT_signal', size=(50))
plt.xlabel('Time [s]', size=(25))
plt.ylabel('SPaT_signal', size=(25))
plt.plot('time_sec', 'SPaT_signal', data=df_red, color = 'red')
plt.plot('time_sec', 'SPaT_signal', data=df_yellow, color = 'yellow')
plt.plot('time_sec', 'SPaT_signal', data=df_green, color = 'green')
```
