```python
import numpy as np
```

```python
def linear(x, Derivative=False):
    if not Derivative:
        return x
    else:
        return 1.0
```

```python
def hardlim(x):
    if x>= 0:
        return 1
    else:
        return 0
```

```python
def sigmoid(x,Derivative=False):
    if not Derivative:
        return 1 / (1 + np.exp (-x))
    else:
        out = sigmoid(x)
        return out * (1 - out)
```

```python
p=2.0
w=2.3
b=-3
z=w*p+b
z
```

```python
linear(z)
```

```python
hardlim(z)
```

```python
sigmoid(z)
```
