A two-input neuron has a bias of 1.2, a weight of [3 2] and an input vector [-5 6]. Draw
the neuron (hand-sketch, Google Drawing, etc. is ok) and calculate the output for a
symmetrical hard limit transfer function, a saturating linear transfer function, and a
hyperbolic tangent sigmoid transfer function in python. Please submit your code and
answer as an *.ipynb file. (3 pts)

```python
import numpy as np
```

```python
bias=1.2
weight=[3, 2]
input_vector=[-5, 6]
z= np.multiply(weight, input_vector).sum() + bias
z 
```

```python
def saturating_linear(x):
    if x< 0:
        return 0
    elif 0<= x<= 1:
        return x
    else:
        return 1    
```

```python
def symmetrical_hard_limit_transfer(x):
    if x>= 0:
        return 1
    else:
        return -1
```

```python
def hyperbolic_tangent_sigmoid_transfer(x):
    return 2 / (1 + np.exp (-2*x)) - 1
```

```python
saturating_linear(z)
```

```python
symmetrical_hard_limit_transfer(z)
```

```python
hyperbolic_tangent_sigmoid_transfer(z)
```
